```python
data_array = []

for stock in stocks:
    # download fundamental data
    data = yf.Ticker(stock)
    fundamentals = data.info
    financials = data.get_financials()
    balance_sheet = data.get_balance_sheet()
    income = data.get_income_stmt()
    try:
        data_array.append(
            {
                "stock": stock,
                "ebit": financials.loc["EBIT"].iloc[0],
                "enterpriseValue": fundamentals["enterpriseValue"],
                "netPPE": balance_sheet.loc["NetPPE"].iloc[0],
                "workingCapital": balance_sheet.loc["WorkingCapital"].iloc[0],
            }
        )
    except Exception as e:
        print(f"Error for {stock}: {e}")
        continue

    time.sleep(1)
```

```python
df = pd.DataFrame(data_array)
df
```

```python
df["EarningsYield"] = df["ebit"] / df["enterpriseValue"]
df["ROIC"] = df["ebit"] / (df["netPPE"] + df["workingCapital"])
df
```

```python
df["EarningsYieldRank"] = df["EarningsYield"].rank(ascending=False)
df["ROICRank"] = df["ROIC"].rank(ascending=False)
df
```

```python
df["CombinedRank"] = df["EarningsYieldRank"] + df["ROICRank"]
df["MagicFormulaRank"] = df["CombinedRank"].rank()
df.sort_values("MagicFormulaRank", inplace=True)
df
```
